```python
plt.figure(figsize = (8,5), facecolor= "LightBlue")
plt.scatter(df["Sorting Time"], df["Delivery Time"])
plt.xlabel("Sorting Time")
plt.ylabel("Delivery Time")
```

```python
df["DT"] = df["Delivery Time"]
df["ST"] = df["Sorting Time"]
```

```python
df
```

```python
model1 = smf.ols("DT ~ ST", data = df).fit()
model1.summary()
```

```python
model2 = smf.ols("np.log(DT) ~ ST", data = df).fit()
model2.summary()
```

```python
model3 = smf.ols("DT ~ np.log(ST)", data = df).fit()
model3.summary()
```

```python
df["ST_sq"] = df["Sorting Time"] ** 2
model4 = smf.ols("np.log(DT) ~ ST ", data = df).fit()
model4.summary()
```

```python
# Out of all Models, Model 4 is more Efficient
```

```python
plt.figure(figsize = (8,5), facecolor= "LightBlue")
plt.scatter(df.ST, df.DT)
plt.scatter(df.ST, np.exp((model4.predict(df.ST))), color = "Red")
plt.plot(df.ST, np.exp((model4.predict(df.ST)), color = "Red")
plt.xlabel("Sorting Time")
plt.ylabel("Delivery Time")
```
